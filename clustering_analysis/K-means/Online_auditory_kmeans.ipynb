## Making necessary file for R code to visualize clusters with bars and individual point

```python
df_aud=pd.read_csv('C:/Users/zhaleh/PhD_Projects/Hierarchy_to_Beat_Perception_Auditory_Visual/results/Three_Auditory_clusters.csv')
cluster1aud=df_aud[df_aud['labels']==0].reset_index()
len(cluster1aud)
```

```python
cluster2aud=df_aud[df_aud['labels']==1].reset_index()
len(cluster2aud)
```

```python
cluster3aud=df_aud[df_aud['labels']==2].reset_index()
len(cluster3aud)
```

```python
########################################################## cluster 1
audC1=pd.Series(np.arange(1,len(cluster1aud)*3+1))
for i in range(0,len(cluster1aud)*3):
    if i < len(cluster1aud):
        audC1[i]='Single duration timing'
    elif i < len(cluster1aud)*2:
        audC1[i]='Sequence timing'
    else:
        audC1[i]='Beat-based timing'
dfC1=pd.concat([cluster1aud['AudSD'],cluster1aud['AudSeq'],cluster1aud['AudBeat']],axis=0,ignore_index=True)

dfCluster1=pd.DataFrame({'Timing.level':audC1,'Cluster.1':dfC1})
########################################################## cluster 2
audC2=pd.Series(np.arange(1,len(cluster2aud)*3+1))
for j in range(0,len(cluster2aud)*3):
    if j < len(cluster2aud):
        audC2[j]='Single duration timing'
    elif j < len(cluster2aud)*2:
        audC2[j]='Sequence timing'
    else:
        audC2[j]='Beat-based timing'
dfC2=pd.concat([cluster2aud['AudSD'],cluster2aud['AudSeq'],cluster2aud['AudBeat']],axis=0,ignore_index=True)

dfCluster2=pd.DataFrame({'Timing.level2':audC2,'Cluster.2':dfC2})
########################################################## cluster 3
audC3=pd.Series(np.arange(1,len(cluster3aud)*3+1))
for j in range(0,len(cluster3aud)*3):
    if j < len(cluster3aud):
        audC3[j]='Single duration timing'
    elif j < len(cluster3aud)*2:
        audC3[j]='Sequence timing'
    else:
        audC3[j]='Beat-based timing'
dfC3=pd.concat([cluster3aud['AudSD'],cluster3aud['AudSeq'],cluster3aud['AudBeat']],axis=0,ignore_index=True)

dfCluster3=pd.DataFrame({'Timing.level3':audC3,'Cluster.3':dfC3})
######################################################### Creating the final file 
final=pd.concat([dfCluster1,dfCluster2,dfCluster3],axis=1)
final.to_csv('C:/Users/zhaleh/PhD_Projects/Hierarchy_to_Beat_Perception_Auditory_Visual/results/Three_Auditory_clusters_2.csv',index=None)
```
