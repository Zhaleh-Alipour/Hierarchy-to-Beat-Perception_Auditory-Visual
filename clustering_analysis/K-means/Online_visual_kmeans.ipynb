## Clusters with all 3 features

```python
df=pd.read_csv('C:/Users/zhaleh/PhD_Projects/Hierarchy_to_Beat_Perception_Auditory_Visual/results/PreprocessedResults_101participants.csv')
df2=df[['VisSD','VisSeq','VisBeat']]
df2.head
```

```python
model2=KMeans(n_clusters=2, init='k-means++',max_iter=1000, random_state=45)
model2.fit(df2)
```

```python
df2['Cluster']=model2.fit_predict(df2)
df2.head()
```

```python
labels=model2.labels_
centroids=model2.cluster_centers_
df2['labels']=labels
df2.head()
```

## Saving the result in the form of a csv file with clusters to which each person belong

```python
df2.drop(columns=['Cluster'],inplace=True)
df2
```

## Making necessary file for R code to visualize clusters with bars and individual point

```python
########################################################## cluster 1
visC1=pd.Series(np.arange(1,196))
for i in range(0,195):
    if i < 65:
        visC1[i]='Single duration timing'
    elif i < 130:
        visC1[i]='Sequence timing'
    else:
        visC1[i]='Beat-based timing'
dfC1=pd.concat([cluster1vis['VisSD'],cluster1vis['VisSeq'],cluster1vis['VisBeat']],axis=0,ignore_index=True)

dfCluster1=pd.DataFrame({'Timing.level':visC1,'Cluster.1':dfC1})
######################################################### cluster 2
visC2=pd.Series(np.arange(1,109))
for j in range(0,108):
    if j < 36:
        visC2[j]='Single duration timing'
    elif j < 72:
        visC2[j]='Sequence timing'
    else:
        visC2[j]='Beat-based timing'
dfC2=pd.concat([cluster2vis['VisSD'],cluster2vis['VisSeq'],cluster2vis['VisBeat']],axis=0,ignore_index=True)

dfCluster2=pd.DataFrame({'Timing.level2':visC2,'Cluster.2':dfC2})
######################################################### creating the final file
final=pd.concat([dfCluster1,dfCluster2],axis=1)
final.to_csv('C:/Users/zhaleh/PhD_Projects/Hierarchy_to_Beat_Perception_Auditory_Visual/results/Two_Visual_clusters_2.csv',index=None)
```
